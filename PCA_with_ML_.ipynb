# Measure Co-variance matrix

```python
import numpy as np

def covariance_matrix(x_sub, y_sub):
    # Create a 2D array from x and y
    data = np.array([x_sub, y_sub])
    
    # Calculate the covariance matrix
    cov_matrix = np.cov(data)
    
    return cov_matrix



cov_matrix = covariance_matrix(x_sub,y_sub)
print("Covariance Matrix:")
print(cov_matrix)
```

```python
eigenvalues=np.linalg.eig(cov_matrix)
```
